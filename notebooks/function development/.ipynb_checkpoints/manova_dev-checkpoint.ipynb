```python
import os
import pandas as pd
from scipy import io
import numpy as np
```

```python
loaded_sp = pd.read_csv('../../data/03_primary/cells_reformated.csv')
```

```python
from scipy.stats import ttest_ind
```

```python
markers = ['mean', 'stdev', 'first', 'second', 'frequency', 'vHold' ]
```

```python
cols_to_test = []

for marker in markers:
    temp_list = [t for t in loaded_sp.columns if (marker in t)]
    cols_to_test.extend(temp_list)
    
len(cols_to_test)
```

```python
loaded_sp.shape
```

```python
df_list=[]
for col in cols_to_test:
    dict_ttest = {}
    for celltype in loaded_sp.cellType.unique():
        celltype_array = loaded_sp.loc[loaded_sp['cellType'] == celltype][col]
        other_array = loaded_sp.loc[loaded_sp['cellType'] != celltype][col]

        stat, p = ttest_ind(other_array, celltype_array, nan_policy='omit', equal_var=False)
        if not isinstance(p, float):
            p=0.5
        dict_ttest[celltype] = p
    temp_df = pd.DataFrame.from_dict(dict_ttest, orient='index')
    temp_df.columns = [col]
    df_list.append(temp_df)

```

```python
stat_significance_df = pd.concat(df_list, axis=1)
```
