```python
import pandas as pd
```

```python
data = pd.read_csv('../../data/04_statistical_analysis/stat_analysis_1v1.csv')
```

```python
data.sort_values(['cell_i',	'cell_j', 'p_value'], inplace=True)
```

```python
data[['feature', 'p_value']].nlargest(columns='p_value', n=3).values
```

```python
df_list = []
for cell_i in data.cell_i.unique():
    for cell_j in data.cell_j.unique():
        temp_df = data.loc[(data.cell_i == cell_i)&(data.cell_j == cell_j)][['feature', 'p_value']].copy()
        df_list.append([cell_i, cell_j, temp_df.nsmallest(columns='p_value', n=3).values])
        
stat_synthesis_df = pd.DataFrame(df_list, columns=['cell_i', 'cell_j', 'differenciators'])
```
